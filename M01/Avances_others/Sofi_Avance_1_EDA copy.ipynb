### 1. Vamos a evaluar cada variable numérica por separado

```python
df_EDA_MIA.select_dtypes('number') #para ver solo columnas numéricas
```

##### **Variable 'edad'**

**BoxPlot**

```python
sns.boxplot(x=df_EDA_MIA['edad'])

plt.show()
```

En el box podemos ver que hay un outlier igual a 300, puesto que es irreal la edad de 300 años vamos a eliminarlo

```python
#Identificamos los outliers concretos

Q1 = df_EDA_MIA['edad'].quantile(0.25)
Q3 = df_EDA_MIA['edad'].quantile(0.75)
IQR = Q3 - Q1

limite_superior = Q3 + 1.5 * IQR
limite_inferior = Q1 - 1.5 * IQR

outliers = df_EDA_MIA[
    (df_EDA_MIA['edad'] < limite_inferior) | 
    (df_EDA_MIA['edad'] > limite_superior)
]

print(f'Q1:{Q1},Q3:{Q3}, IQR: {IQR}')
print(f'Límite inferior:{limite_inferior:.2f}),Límite superior: {limite_superior:.2f}\n')

```

```python
# Eliminanos ese outlier

import numpy as np

df_EDA_MIA['edad'] = df_EDA_MIA['edad'].apply(lambda x: x if (x is not None and 0 < x < 112) else np.nan)

sns.boxplot(x=df_EDA_MIA['edad'])

plt.show()
```

##### **Variable frecuencia_visita**

**BoxPlot**

```python
#Vamos a usar BoxPlot para verificar si hay outliers en la columna frecuencia_visita


sns.boxplot(x=df_EDA_MIA['frecuencia_visita'])
plt.show()
```

En el gráfico BoxPlot de frecuencia_visita observamos outliers negativos, estos valores son irreales y vamos a proceder a eliminarlos.

### 2. Vamos a evaluar si existe correlación entre las variables numéricas (Pearson)

```python
pearson = df_EDA_MIA.select_dtypes('number').corr() #Aplicamos la correlación de Pearson

pearson
```

```python
#Graficamos con un Mapa de calor

sns.heatmap(pearson, annot=True, cmap="coolwarm")
plt.title("Mapa de calor de correlaciones")
plt.show()

```

Correlaciones entre variables

- edad con otras variables:
  - Todas las correlaciones son cercanas a 0 (-0.013, -0.003, -0.015), lo que indica prácticamente ninguna relación lineal con las demás variables.

- frecuencia_visita con promedio_gasto_comida y ingresos_mensuales:
  - frecuencia_visita  y promedio_gasto_comida → 0.53 indica una correlación moderada positiva.
  - frecuencia_visita  y ingresos_mensuales → 0.75 indica una correlación fuerte positiva. Esto indica que a mayor número de ingresos_mensuales, mayor es el valor de frecuencia_visita.

- promedio_gasto_comida con ingresos_mensuales:
  - 0.63 → indica correlación moderada-alta positiva. 

Conclusiones principales

1. La edad no parece influir en las demás variables (correlación cercana a 0).

2. frecuencia_visita está fuertemente relacionado con los ingresos_mensuales, y moderadamente con promedio_gasto_comida.

3. promedio_gasto_comida también tiene una correlación moderada con los ingresos_mensuales, lo que podría indicar que mejores salarios generan más gasto.

4. Las variables frecuencia_visita, promedio_gasto_comida e ingresos_mensuales están más relacionadas entre sí que con edad.
