```python
# importar las librerías necesarias
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://en.wikipedia.org/wiki/San_Diego" # URL de la página de Wikipedia de San Diego.

# Configurar el User-Agent para evitar bloqueos
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser") # se parsea el contenido HTML de la página.

# Palabras clave que indican lugares históricos o monumentos
keywords = [
    "historic", "landmark", "mission", "museum", "park", "memorial",
    "monument", "heritage", "preserve", "district", "site", "plaza"
]

historic_places = []

# Recorremos todas las listas
for ul in soup.find_all(["ul", "ol"]): # buscamos todas las listas (ul y ol) cuya función es agrupar elementos relacionados.
    for li in ul.find_all("li"): # buscamos todos los elementos de lista (li) dentro de cada lista (ul o ol).
        text = li.get_text().strip() # obtenemos el texto y eliminamos espacios en blanco.

        # Si el texto contiene alguna palabra clave se guarda
        if any(word in text.lower() for word in keywords):
            historic_places.append(text)

# Eliminación de duplicados
historic_places = list(dict.fromkeys(historic_places))

# Guardar
df = pd.DataFrame({"Historic or Notable Place": historic_places})
df.to_csv("san_diego_historic_places.csv", index=False)

print(df)
print("Guardado como san_diego_historic_places.csv") 

print("Generado por ChatGPT") # imprimo mensaje de la IA que generó el código
```

##### Buscar otra ciudad. Buscar "Del Mar":

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Configuración del driver de Selenium
chrome_options = Options()
chrome_options.add_argument("--start-maximized")

# Inicializar el driver
driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=chrome_options
)

wait = WebDriverWait(driver, 10) # espera explícita de hasta 10 segundos.

driver.get("https://en.wikipedia.org/wiki/San_Diego")
time.sleep(2) # esperar a que la página cargue completamente.

# buscador para buscar otra ciudad

ciudad_buscar = "Del Mar, California" # ciudad a buscar en Wikipedia (agrego una coma y el estado para mayor precisión).

search_box = wait.until(EC.presence_of_element_located((By.ID, "searchInput"))) # caja de búsqueda con id "searchInput".
search_box.clear() # limpiar la caja de búsqueda.
search_box.send_keys(ciudad_buscar) # escribir la ciudad a buscar.
search_box.send_keys(Keys.ENTER) # presionar Enter para buscar.

time.sleep(3) # esperar a que la página de resultados cargue.
```

###### Al ejecutar el scrpit de arriba, se abrirá Wikipedia en la ciudad de San Diego y luego el buscador escribirá "Del Mar" en la caja de búsqueda y nos llevará a la ciudad de Del Mar, California.

##### Cambiar el tema a Dark y de idioma al italiano:

```python
# cambiar el tema a Dark
try:
    theme_button = wait.until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "button#vector-page-theme-toggle"))
    ) # botón para cambiar el tema con id "vector-page-theme-toggle".
    theme_button.click() # hacer clic en el botón para cambiar el tema.
    time.sleep(2) # esperar a que el tema cambie.
except:
    print("No se encontró el botón de tema. Puede haber cambiado el selector.") # mensaje de advertencia si no se encuentra el botón.

# cambiar el idioma de la página al italiano
try:
    # Abrir el menú de idiomas
    lang_button = wait.until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "button#p-lang-btn")) 
    ) # botón de idiomas con id "p-lang-btn".
    lang_button.click()
    time.sleep(2)

    # Elegir un idioma (ejemplo: italiano)
    lang_link = wait.until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "a[lang='it']")) # enlace al idioma italiano con atributo lang='it'.
    )
    lang_link.click() # hacer clic en el enlace del idioma italiano.

    time.sleep(4) # esperar a que la página en italiano cargue.

except:
    print("No se pudo cambiar el idioma. Puede que no exista ese idioma para esta página.") # mensaje de advertencia si no se puede cambiar el idioma.

print("Automatización con Selenium completada.") # mensaje de finalización.
time.sleep(3) # esperar antes de cerrar el navegador.
driver.quit() # cerrar el navegador.
```
