# For y While

```python
suma=0
suma=suma+1
print(suma)
```

```python
suma=0
suma=suma+1
suma=suma+2
print(suma)
```

```python
numeros=[1,2,3,4,5]
for num in numeros:
    suma=num+1
    print(suma)
```

```python
for i in range (1,6):
    suma=i+1
    print(suma)
```

```python
for i in range (1,6):
    suma=i+1
    print("Variable de iteración:", i, " Resultado=", suma)
```

```python
num=1
while num<6:
    print("num", num)
    num=num+1
```

```python
for i in range(5):
    print(i)
    break
```

```python
cadena="Python"
for letra in cadena:
    if letra=="h":
        print("Se encontró la h")
        break
    print(letra)
```

```python
x=5
while True:
    x -= 1
    print(x)
    if x==0:
        break
    print("Fin del bucle")
```

```python
cadena="Python"
for letra in cadena:
    if letra=="P":
        continue
    print(letra)
```
