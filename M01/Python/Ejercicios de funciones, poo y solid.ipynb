7. Principio SRP (responsabilidad única)

Corrigí este código para cumplir SRP:

class Reporte:
    def generar(self):
        print("Generando reporte")
    def guardar(self):
        print("Guardando en disco")
    def enviar_email(self):
        print("Enviando por email")


Separalo en tres clases con responsabilidades claras.
