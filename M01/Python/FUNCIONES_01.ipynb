```python
def semana(dia): # dia entre paréntesis es el argumento
    fin_de_semana = ( 'sabado', 'domingo') # tupla
    dia_laboral = ('lunes', 'martes', 'miercoles', 'jueves', 'viernes') # tupla
    if dia in fin_de_semana: # "if" siendo el condicional
        print('Fin de semana')
    else:
        print('Dia laboral')
 
    return 
semana('sabado')
```

```python
# DOCSTRINGS

def greet(name: str) -> str:
    """
    Devuelve un saludo personalizado.

    Parámetros:
    name (str): El nombre de la persona a saludar.

    Retorna:
    str: Un saludo personalizado.
    """
    return f"Hola, {name}!"
```

```python
# TIPADO DE DATOS

def suma(a: int, b: int) -> int:
    """
    Suma dos números enteros.

    Parámetros:
    a (int): El primer número.
    b (int): El segundo número.

    Retorna:
    int: La suma de a y b.
    """
    return a + b
```
