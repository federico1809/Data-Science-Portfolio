```python
class Gato:
    
    def __init__(self,nombre,edad,raza,sexo):
        self.nombre = nombre
        self.edad = edad
        self.raza = raza
        self.sexo = sexo
    
# Método Presentar

    def presentar(self):
        return print('-Mi Nombre:',self.nombre,'-Edad:',self.edad,'-Raza:',self.raza,'-Sexo:',self.sexo)
```

```python
gato1 = Gato('Atena','2 años','calico','hembra')
```

```python
print(gato1)
```

```python
gato1.presentar()
```

```python
gato2 = Gato('Pancha','1 año', 'calico', 'hembra')
```

```python
gato2.presentar()
```

```python
class Animal:
    def __init__(self,especie,edad,color):
        self.especie = especie
        self.edad = edad
        self.color = color

    def me_presento(self):
        print('Hola, soy un',self.especie,'de color',self.color,'y tengo',self.edad,'año/s de edad.')

    def cumplirAños(self):
        self.edad = self.edad + 1
```

```python
a1 = Animal('Ratón',1,'Marrón')
a2 = Animal('Liebre',2,'Gris')
```

```python
print(a1.especie)
print(a1.edad)
print(a1.color)
print(a2.especie)
print(a2.edad)
print(a2.color)
```

```python
a1.me_presento()
a2.me_presento()
```

```python
a1.cumplirAños()
a2.cumplirAños()
```

```python
class CuentaBancaria:
    def __init__(self,saldo):
        self.__saldo = saldo # El atributo saldo está encapsulado con __

    def depositar(self, monto):
        self.__saldo += monto
    
    def retirar(self, monto):
        if self.__saldo >= monto:
            self.__saldo -= monto
        else:
            print('Saldo insuficiente')
    
    def obtener_saldo(self):
        return self.__saldo
```

```python
class Mensaje:
    def enviar(self, mensaje):
        pass

class SMS(Mensaje):
    def enviar(self, mensaje):
        print("Enviando SMS:", mensaje)

class Usuario:
    def __init__(self, medio_mensaje):
        self.medio_mensaje = medio_mensaje

    def enviar_mensaje(self, mensaje):
        self.medio_mensaje.enviar(mensaje)

# Crear una instancia de SMS
sms = SMS()

# Crear una instancia de Usuario que utiliza SMS para enviar mensajes
usuario = Usuario(sms)

# Enviar un mensaje
usuario.enviar_mensaje("Hola, este es un mensaje de prueba")
```
