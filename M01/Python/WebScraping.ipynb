# Web Scraping

```python
# importamos las librerías necesarias
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

### Creamos un objeto Response a partir de una página HTML y salvaremos el resultado en una variable que llamaremos soup:

```python
# Creamos un objeto Response a partir de una página HTML y salvaremos el resultado en una variable que llamaremos response
URL = 'https://store.data-analyst.praktikum-services.ru/en/'

response = requests.get(URL)
```

```python
# busco el tipo de dato que tiene el atributo text del objeto response
type(response.text)
```

```python
# obtengo el contenido HTML de la página
response.text
```

```python
response.status_code # 200 significa que la página se ha cargado correctamente
```

### Obtenemos los nombres de los productos a partir de la URL (ya la tenemos en el objeto response) y la vamos a añadir a una lista nombre_productos

```python
soup = BeautifulSoup(response.text, 'lxml')
#creamos la lista donde guardaremos los nombres de los productos
precio = []

for row in soup.find_all('div', attrs = {'class': 't754__price-value js-product-price'}):
    precio.append(row.text)

print(precio)
```

### Construimos el dataframe


```python
# extraigo los nombres de los productos y los guardo en una lista
nombre_productos = []
for row in soup.find_all('div', attrs={'class': 't754__title t-name t-name_md js-product-name'}):
	nombre_productos.append(row.text)

# creamos el DataFrame con los datos obtenidos
productos_data = pd.DataFrame({
	'Nombre': nombre_productos,
	'Precio': precio
})

productos_data.head(10)

```
