Funciones Lambda
1) Verificación de Entrega

Endpoint sugerido: POST /delivery/verify

Consulta en DynamoDB el estado actual.

2) Cálculo de ETA

Trigger sugerido: EventBridge

Calcula hora estimada de llegada usando distancia restante.

3) Alerta por Desvío

Trigger sugerido: IoT Core / GPS stream

Envía alerta SNS

Registra incidente en DynamoDB

Usa cálculo geográfico Haversine

Ejemplo de Lambda en Python

```python
import boto3
import json

dynamodb = boto3.resource('dynamodb')
table = dynamodb.Table('Deliveries')

def lambda_handler(event, context):
    delivery_id = event['delivery_id']
    response = table.get_item(Key={'delivery_id': delivery_id})
    return {
        'statusCode': 200,
        'body': json.dumps(response.get('Item', {}))
    }
```

Flujo de Datos

API Gateway → Lambda

Lambda → DynamoDB / SNS / S3

RDS actúa como base transaccional migrada desde local
